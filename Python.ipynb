```python
# Set operations
# Union()- combines two sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1.union(set2)
print(result)
```

```python
# Alternatively, use the | operator,
result = set1 | set2
print(result)
```

```python
# Intersection 
set1 = {1, 2, 3, 4}
set = {3, 4, 5, 6}
result = set1.intersection(set2)
print(result)
```

```python
# Alternatively, use the & operator
rsult = set1 & set2
print(result)
```

```python
# Difference ()- elements in one set but not the other
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set2.difference(set1)
print(result)
result = set1.difference(set2)
print(result)
```

```python
# Alternatively, use the - operator
result = set1 - set2
print(result)
result = set2 - set1
print(result)
```

```python
# Symmetric difference
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set1.symmetric_difference(set2)
print(result)

# Alternatively, use the ^ operator
result = set1 ^ set2
print(result)
```

```python
# issubset - check if one set is a subset of another
set1 = {1, 2}
set2 = {1, 2, 3, 4}
print(set1.issubset(set2))

# Alternatively, use <=
print(set1 <= set2)
```

```python
# issuperset
set1 = {1, 2, 3, 4}
set2 = {1, 2}
print(set1.issuperset(set2))

# Alternatively, use >= 
print(set1 >= set2)
```

```python
# isdisjoint 
set1 = {1, 2, 3}
set2 = {4, 5, 6}
print(set1.intersection(set2))

print(set1.isdisjoint(set2))
```

```python
# Bitwise Operations
a = 5 # Binary: 101
b = 3 # Binary: 011
print(a & b) # 1 (Binary: 001)
print(a | b) # 7 (Binary: 111)
print(a ^ b) # 6 (Binary: 110)
print(~a) # -6 (Binary: -(101+1))
print(a << 1) # 10 (Binary: 1010)
print(a >> 1) # 2 (Binary: 10)
```
