1: Working with Set - 10 points
Two groups of students enrolled in different courses.
Inputs:
python = {"Adeola", "Zainab", "Funmi", "Tomi", "Abiola", "Nofisat"}
ml = {"Abiola", "Halimat", "Funmi", "Tomi", "Arafat"}

a). Find students in both courses
b). Find all students
c). Find students only in Python

```python
python = {"Adeola", "Zainab", "Funmi", "Tomi", "Abiola", "Nofisat"}
ml = {"Abiola", "Halimat", "Funmi", "Tomi", "Arafat"}

result = python & ml #Intersection of python and ml
print(result)

result = python | ml #Union of python and ml
print(result)

print(python)
```

2: Website Visitor Tracker - 10 points
You are tracking unique visitors to a website. Given the names of the visitors ['Abiola', 'Halimah', 'Zainab', 'Esther', 'Lola', 'Carina', 'Salamat', 'Funmi', 'Zainab', 'Jamilah', 'Lola', 'Hannah', 'Esther'].

Print all visitors
Print the unique visitors in alphabetical order

```python
visitors = ['Abiola', 'Halimah', 'Zainab', 'Esther', 'Lola', 'Carina', 'Salamat', 'Funmi', 'Zainab', 'Jamilah', 'Lola', 'Hannah', 'Esther']
print(visitors)

unique_visitors = (sorted.(visitors))
print(unique_visitors)
```
